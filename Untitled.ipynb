```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
s=np.linspace(1,100,1000).astype(float)
Re=np.array([i*10**(4) for i in s])
Ne_1D=lambda x: x**(3/4)
Ne_3D=lambda x: x**(2.25)
```

```python
N_1D=N_1D(Re)
N_3D=N_3D(Re)
```

```python
plt.plot(Re,N_1D*100000000,label='1D factored to 9th power to match scale')
plt.plot(Re,N_3D,label='3D')
plt.xlabel('Re')
plt.ylabel('Number of points')
plt.title('Mesh size dependence')
plt.legend()
plt.show()
```

```python
plt.plot(np.log(Re),np.log(N_1D),label='1D')
plt.plot(np.log(Re),np.log(N_3D),label='3D')
plt.xlabel('log(Re)')
plt.ylabel('log(N)')
plt.title('Mesh size dependence log scale')
plt.legend()
plt.show()
```

```python
Msh_in_units=N_3D/(np.min(N_3D))
```

```python
#plt.plot(Re,N_1D*100000000,label='1D factored to 9th power to match scale')
plt.plot(Re,Msh_in_units,label='3D')
plt.xlabel('Re')
plt.ylabel('Cost')
plt.title('Cost vs Re')
plt.legend()
plt.show()
```

```python
#plt.plot(Re,N_1D*100000000,label='1D factored to 9th power to match scale')
plt.plot(np.log(Re),np.log(Msh_in_units),label='3D')
plt.xlabel('log(Re)')
plt.ylabel('log(Cost)')
plt.title('Cost vs Re')
plt.legend()
plt.show()
```
