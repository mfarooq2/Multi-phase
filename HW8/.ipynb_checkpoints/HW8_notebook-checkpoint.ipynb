```python
import numpy as np
import math
import numpy.linalg as LA
import matplotlib.pyplot as plt
```

```python
rho_l=999.21
rho_g=1.205
g=9.8
C_d=0.49
u_l=1e-3
gamma=7.27e-2
D_b=1e-3*np.linspace(0.5,30,30)
# Vol_b=(1/6)*np.pi*(D_b)**3
# A_x_b=(1/4)*np.pi*(D_b)**2
m_b=rho_g*Vol_b
v_r_c=np.sqrt((4*(rho_l-rho_g)*(D_b)*g)/(3*C_d*rho_l))
```

```python
plt.plot(D_b,v_r,label='Dia vs Rise_Velocity')
plt.xlabel('D (m)')
plt.ylabel('$V_r$ (m/s)')
plt.legend()
plt.show()
```

```python
def C_d_Func(rho_l,D_b,u_l,gamma,v_r,rho_g,g):
    Re_b=(rho_l*D_b*v_r)/(u_l)
    E_o=(rho_l-rho_g)*g*((D_b)**2)/gamma
    C_d_v=np.sqrt(((16/Re_b)*(1+(2)/(1+(16/Re_b)+(3.315/Re_b**0.5))))**2+((4*E_o)/(E_o+9.5))**2)
    return Re_b,E_o,C_d_v
```

```python
for i in range(50):
    Re_b,E_o,C_d_v=C_d_Func(rho_l,D_b,u_l,gamma,v_r,rho_g,g)
    v_rn=np.sqrt((4*(rho_l-rho_g)*(D_b)*g)/(3*(C_d_v)*rho_l))
    LA.norm(v_rn-v_r)
    v_r=v_rn.copy()
```

```python
plt.plot(D_b,v_rn,label='Variable Drag Coefficient')
plt.plot(D_b,v_r_c,label='Constant Drag Coefficient')
plt.xlabel('D (m)')
plt.ylabel('$V_r$ (m/s)')
plt.legend()
plt.show()
```
