```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
s=np.linspace(0,100,1000).astype(float)
Re=np.array([i*10**(4) for i in s])
N_1D=lambda x: x**(3/4)
N_3D=lambda x: x**(2.25)
```

```python
plt.plot(Re,N_1D(Re)*100000000,label='1D factored to 9th power to match scale')
plt.plot(Re,N_3D(Re),label='3D')
plt.xlabel('Re')
plt.ylabel('Number of points')
plt.title('')
plt.legend()
plt.show()
```
